```python
#This script downloads a FITS image from the Legacy Survey Viewer based on specified coordinates, size, band, and catalog version.

import urllib.request

# Exact coordinates of the center
ra = 199.107905
dec = 17.6822745

# Image size in pixels.
# 3000 pixels at 0.262 arcsec/pix = ~13 minutes of arc (covers the entire cluster)
size = 3000 

# Band (g, r, z) and version of the catalog (dr10 is the current standard)
banda = 'g'
layer = 'ls-dr10'

# Construct the API URL
url = f"https://www.legacysurvey.org/viewer/fits-cutout?ra={ra}&dec={dec}&layer={layer}&pixscale=0.262&bands={banda}&size={size}"
nombre_archivo = f"DESI_{layer}_banda_{banda}.fits"

try:
    # Download and save the FITS file
    urllib.request.urlretrieve(url, nombre_archivo)
    print(f"[+] Image saved as: {nombre_archivo}")
except Exception as e:
    print(f"[-] Error downloading the image: {e}")
```
